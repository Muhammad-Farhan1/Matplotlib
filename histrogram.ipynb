Histogram in Matplotlib is a graphical representation that organizes a group of data points into user-specified ranges. Similar to a bar chart, it displays the frequency of data points that fall within each range, known as bins. Histograms are particularly useful for visualizing the distribution of numerical data, allowing you to see patterns such as skewness, modality, and the presence of outliers.

```python
import matplotlib.pyplot as plt

ages = [10 ,34, 45, 32, 54, 34, 67, 55, 43, 32, 56,59,11,23,67,45,66,60,65]

plt.hist(ages , bins=6 , color='lightpink' , edgecolor='black' , width=1.4)
plt.xlabel('Ages of peeps')
plt.ylabel('Number of peeps')
plt.title('peeps with different age')
plt.show()
```

```python
# vertical histogram
plt.hist(ages , bins=6 , orientation='horizontal' , color='lightblue' , edgecolor='black' , height=1.4)
plt.ylabel('Ages of peeps')
plt.xlabel('Number of peeps')
plt.title('peeps with different age')
plt.show()
```
